```python
import scipy.stats as stats

def one_way_anova(groups):
    f_statistic, p_value = stats.f_oneway(*groups)

    return f_statistic, p_value
group1 = [35, 38, 42, 28, 47]
group2 = [40, 45, 50, 55, 58]
group3 = [25, 30, 35, 20, 28]

f_statistic, p_value = one_way_anova([group1, group2, group3])
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

```python
import scipy.stats as stats
columns_to_investigate = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']
grouped = white_wine.groupby('quality_category')
print('White wine')
#  calculate ANOVA for columns
for column in columns_to_investigate:
    print (column)
    # Create a dictionary to store the data for each quality category
    category_data = {}
    for category, group_data in grouped:
        category_data[category] = group_data[column]
    # Perform ANOVA
    f_statistic, p_value = stats.f_oneway(*category_data.values())
    print(f"F-statistic: {f_statistic}, P-value: {p_value}")
```

```python
#Relative Importance Index for white wine

#used the same as before columns_to_investigate
columns_to_investigate = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']

white_wine['quality_normalized'] = (white_wine['quality'] - white_wine['quality'].min()) / (white_wine['quality'].max() - white_wine['quality'].min())
rii = {}
for col in columns_to_investigate:
    white_wine[col + '_normalized'] = (white_wine[col] - white_wine[col].min()) / (white_wine[col].max() - white_wine[col].min())
    rii[col] = (white_wine[col + '_normalized'] * white_wine['quality_normalized']).sum() / white_wine['quality_normalized'].sum() * 100
rii_df = pd.DataFrame(rii.items(), columns=['Factor', 'RII'])
rii_ranking = rii_df.sort_values(by='RII', ascending=False)
display(rii_ranking)
```

```python
#Relative Importance Index for red wine
red_wine['quality_normalized'] = (red_wine['quality'] - red_wine['quality'].min()) / (red_wine['quality'].max() - red_wine['quality'].min())
rii = {}
for col in columns_to_investigate:
    red_wine[col + '_normalized'] = (red_wine[col] - red_wine[col].min()) / (red_wine[col].max() - red_wine[col].min())
    rii[col] = (red_wine[col + '_normalized'] * red_wine['quality_normalized']).sum() / red_wine['quality_normalized'].sum() * 100
rii_df = pd.DataFrame(rii.items(), columns=['Factor', 'RII'])
rii_ranking = rii_df.sort_values(by='RII', ascending=False)
display(rii_ranking)
```

```python
columns_to_investigate = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

def create_boxplots(data, x_col, y_cols):
    """
    Create boxplots for multiple columns in a wine dataFrame.
    Parameters:
    - data: Name of DataFrame (red_wine or whitewine in this case)
    - x_col: Column to use on the x-axis (e.g., 'quality_category')
    - y_cols: List of columns to create boxplots for (e.g., ['fixed acidity', 'volatile acidity'])
    - or just define columns_to_investigate and call them as y_cols
    """
    for y_col in y_cols:
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=data, x=x_col, y=y_col)
        plt.title(f'Boxplot of {y_col} by {x_col}')
        plt.xlabel(x_col)
        plt.ylabel(y_col)
        plt.show()

print('White wine Boxplots')
create_boxplots(white_wine, 'quality_category', columns_to_investigate)
```
