```python
#marking outliers red_wine
red_wine['outliers'] = np.nan
#outliers with total sulfur dioxide in red wine (more then 160)
red_wine.loc[red_wine['total sulfur dioxide'] > 160, 'outliers'] = 'total sulfur dioxide too high'
red_wine.loc[red_wine['volatile acidity'] > 1.2, 'outliers'] = 'max4 of volatile acidity'
red_wine.loc[red_wine['residual sugar'] > 15.0, 'outliers'] = 'max3 of residual sugar'
red_wine.loc[red_wine['alcohol'] > 14.0, 'outliers'] = 'max of alcohol'
red_wine.loc[red_wine['pH'] < 2.9, 'outliers'] = 'lower_9 Ph'
red_wine.loc[red_wine['pH'] > 4, 'outliers'] = 'highest_2 Ph'

red_wine.head()
```

```python
#marking outliers white_wine
white_wine['outliers'] = np.nan
white_wine.head()
white_wine.loc[white_wine['quality'] == 9 , 'outliers'] = 'top 5 guality'
white_wine.loc[white_wine['volatile acidity'] > 0.95 , 'outliers'] = 'top3 vol.acidity in WW'
white_wine.loc[white_wine['citric acid'] > 1 , 'outliers'] = 'top2 citric acid'
white_wine.loc[white_wine['citric acid'] > 1 , 'outliers'] = 'top3 vol.acidity in WW&'
white_wine.loc[white_wine['total sulfur dioxide'] > 300 , 'outliers'] = 'total sulfur dioxide >300'
white_wine.loc[white_wine['sulphates'] >= 0.99 , 'outliers'] ='top5 sulphates'
white_wine.loc[white_wine['alcohol'] >= 14.00 , 'outliers'] = 'top7 alcohol'
white_wine.loc[white_wine['alcohol'] == 8.00 , 'outliers'] = 'min2 alcohol'
```

```python
# Creating a column 'quality wine' and dividing into 3 categories
cond_rw = [
    red_wine['quality'] > 7,
    red_wine['quality'] > 5,
    red_wine['quality'] > 1]
values = [
    'higher quality',
    'medium quality',
    'lower quality']
red_wine['quality_category'] = np.select(cond_rw, values, default=np.nan)

cond_ww = [
    white_wine['quality'] > 7,
    white_wine['quality'] > 5,
    white_wine['quality'] > 1]
white_wine['quality_category'] = np.select(cond_ww, values, default=np.nan)

#another way to do same with lambda 
# red_wine['quality_category'] = red_wine['quality'].apply(lambda value: 'lower quality'
# if value <= 5 else 'medium quality'
# if value < 7 else 'higher quality')

# here we are transforming these labels into categrical data type (specific to pandas) instead of simple string
red_wine['quality_category'] = pd.Categorical(red_wine['quality_category'],
categories=['lower quality', 'medium quality', 'higher quality'])
white_wine['quality_category'] = pd.Categorical(white_wine['quality_category'],
categories=['lower quality', 'medium quality', 'higher quality'])
```

```python
hist1 = sns.histplot(data=red_wine, x='quality_category', color="darkred")
#plt.ylim(0, 3000)
# plt.bar('Quality Category', 'Count')
plt.xlabel('Quality Category')
plt.ylabel('Count')
plt.title('Distribution of Quality Categories in Red Wine')
```

```python
hist2 = sns.histplot(data=white_wine, x='quality_category', color="lightyellow")
# plt.bar('Quality Category', 'Count')
plt.xlabel('Quality Category')
plt.ylabel('Count')
plt.title('Distribution of Quality Categories in White Wine')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# The first histogram
sns.histplot(data=red_wine, x='quality_category', color="darkred", ax=ax1)
ax1.set_xlabel('Quality Category')
ax1.set_ylabel('Count')
ax1.set_title('Distribution of Quality Categories in Red Wine')

# The second histogram
sns.histplot(data=white_wine, x='quality_category', color="lightyellow", ax=ax2)
ax2.set_xlabel('Quality Category')
ax2.set_ylabel('Count')
ax2.set_title('Distribution of Quality Categories in White Wine')
ax1.set_ylim(0, 3200) 

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.15, hspace=0.6)
plt.show()
```

```python
labels = 'Red wines samples', 'White wines samples'
samples = [red_wine['wine_type'].count(), white_wine['wine_type'].count()]
explode = (0.1, 0)  
colors = ['darkred', 'lightyellow']
fig1, ax1 = plt.subplots()
ax1.pie(samples, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, colors=colors)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# boxplot for Acohol&pH in Red Wines
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
# First boxplot for 'alcohol' by 'quality_category' for redwine
sns.boxplot(data=red_wine, x = 'quality_category', y = 'alcohol', ax=ax1)
ax1.set_title('Alcohol Content vs. Quality Category in Red wine samples')
ax1.set_xlabel('Quality Category')
ax1.set_ylabel('Alcohol Content')
# Second for white wine
sns.boxplot(data=red_wine, x='quality_category', y = 'pH', ax=ax2)
ax2.set_title('pH level vs. Quality Category in Red wine samples')
ax2.set_xlabel('Quality Category')
ax2.set_ylabel('pH level')

plt.subplots_adjust(wspace=0.4)
plt.show()
```

```python
# boxplot for Acohol&pH in White Wines
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
# First boxplot for 'alcohol' by 'quality_category' for white wine
sns.boxplot(data=white_wine, x='quality_category', y='alcohol', ax=ax1)
ax1.set_title('Alcohol Content vs. Quality Category in White wine samples')
ax1.set_xlabel('Quality Category')
ax1.set_ylabel('Alcohol Content')
# Second boxplot for 'pH' by 'quality_category' for white wine
sns.boxplot(data=white_wine, x='quality_category', y='pH', ax=ax2)
ax2.set_title('pH level vs. Quality Category in White wine samples')
ax2.set_xlabel('Quality Category')
ax2.set_ylabel('pH level')

plt.subplots_adjust(wspace=0.2)
plt.show()
```

```python
red_wine_num = red_wine.select_dtypes(include=['int', 'float'])
corr_matrix = red_wine_num.corr()
plt.figure(figsize=(14, 8))
sns.heatmap(corr_matrix, annot=True, cmap='bwr', vmin=-1, vmax=1) #seismic
plt.suptitle("Correlation coefficients between features in the Red wine samples", y=0.93, x=0.41)
plt.show()
```
